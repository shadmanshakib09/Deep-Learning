```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Number of data points
n = 100

# Generate random humidity and temperature values for two classes
humidity_class_0 = np.random.uniform(10, 30, n//2)
temperature_class_0 = np.random.uniform(60, 80, n//2)

humidity_class_1 = np.random.uniform(30, 50, n//2)
temperature_class_1 = np.random.uniform(80, 100, n//2)

# Create the labels
rain_class_0 = np.zeros(n//2)
rain_class_1 = np.ones(n//2)

# Combine the classes
humidity = np.concatenate([humidity_class_0, humidity_class_1])
temperature = np.concatenate([temperature_class_0, temperature_class_1])
rain = np.concatenate([rain_class_0, rain_class_1])

# Create a DataFrame
df = pd.DataFrame({
    'humidity': humidity,
    'temperature': temperature,
    'rain': rain.astype(int)
})
```

```python
df
```

```python
sns.scatterplot(data=df, x='humidity', y='temperature', hue='rain')
```

```python
X = df.drop('rain', axis=1)
y = df['rain']
```

```python
X.head()
```

```python
y.shape
```

```python
from sklearn.linear_model import Perceptron
```

```python
p = Perceptron(random_state=1)
p.fit(X, y)
```

```python
p.score(X, y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values, clf=p, legend = 2)
```
