```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply the encoding to the 'species' column
df['species_encoded'] = label_encoder.fit_transform(df['species'])

# Display the encoded species values
print("\nEncoded species values:")
print(df[['species', 'species_encoded']].drop_duplicates())
```

```python
df = df.iloc[:,2:] #we will work with petal values
```

```python
df.head()
```

```python
# Drop the 'species' column
df = df.drop('species', axis=1)
df.head()
```

```python
sns.scatterplot(x='petal_length',y='petal_width',hue='species_encoded',data=df)
```

```python
X = df.drop('species_encoded',axis=1)
y = df['species_encoded']
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
clf = LogisticRegression()
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
plot_decision_regions(X_train,y_train.values,clf = clf, legend =2 )
plt.show()
```
