## Read CSV from URL

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

## CONVERTORS

```python
def rename(name):
  if name == 'az':
    return 'New'
  else:
    return name
```

```python
rename('az')
```

```python
pd.read_csv(file, converters={'X0':rename}) #rename any value in column
```

## na values

```python
pd.read_csv(file, na_values=['k']) #make any value converted into NaN
```

```python
dfs = pd.read_csv(file, chunksize=1000)
```
